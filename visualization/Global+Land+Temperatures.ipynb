```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./GlobalLandTemperaturesByCountry.csv')
data['dt'] = pd.to_datetime( data['dt'], format='%Y-%d-%m', errors='ignore' )
data = data[ data['AverageTemperature'].notnull() ]
```

```python
data = data[ data['Country'] == 'Iran' ]
```

```python
data.tail(10)
```

```python
g = sns.FacetGrid(data, size=10)
g.map(plt.scatter, "dt", "AverageTemperature", s=30, alpha=.6)
g.add_legend();
```
